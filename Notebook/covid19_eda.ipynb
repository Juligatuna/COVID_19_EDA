```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("owid-covid-data.csv")
df.sample(10)
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
cols = ['location', 'date', 'total_cases', 'new_cases', 
        'total_deaths', 'new_deaths',
        'total_vaccinations', 'people_fully_vaccinated', 
        'population']
df = df[cols]
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
df['location'].unique()
```

```python
df = df[~df['location'].isin(['Upper middle income', 'Europe', 'European Union', 'High income', 'Lower middle income',
                              'South America','World', 'Oceania', 'North America', 'Africa', 'Asia' ])]
```

# ðŸ”¹ Step 3: Exploratory Data Analysis (EDA)
## 1. Global summary

```python
df.groupby('date')[['new_cases', 'new_deaths']].sum().plot(title= 'Global Covid 19 cases and deaths')
```

```python
import matplotlib.pyplot as plt
global_trend = df.groupby('date')[['new_cases', 'new_deaths']].sum()
```

```python
fig, ax1 = plt.subplots(figsize=(12,5))

ax1.plot(global_trend.index, global_trend["new_cases"], color="blue", label="New Cases")
ax1.set_ylabel("New Cases(in millions)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")


ax2 = ax1.twinx()
ax2.plot(global_trend.index, global_trend["new_deaths"], color="red", label="New Deaths")
ax2.set_ylabel("New Deaths", color="red")
ax2.tick_params(axis="y", labelcolor="red")

fig.suptitle("Global COVID-19 Cases & Deaths")
fig.tight_layout()
plt.show()
```

## 2. Top 10 countries by total cases

```python
top_cases = df.groupby('location')['total_cases'].max().sort_values(ascending = False).head(10)
ax = top_cases.plot(kind='bar', title='Top 10 countries by Total Cases', figsize=(10,6))
ax.set_ylabel("Total Cases(in hundreds of millions)")
```

## 3. Vaccination progress

```python
vacc = df.groupby('location')['people_fully_vaccinated'].max().sort_values(ascending = False).head(10)
ax = vacc.plot(kind = 'bar', title = 'Top 10 countries by fully vaccinated people')
ax.set_ylabel('Total vaccinations(in billions)')
```

```python
df.total_deaths.sum()# the sum is cumulative per day
```

```python
latest_per_country = df.sort_values("date").groupby("location").tail(1)

global_cases = latest_per_country["total_cases"].sum()
global_deaths = latest_per_country["total_deaths"].sum()

print("Global Total Cases:", f"{global_cases:,.0f}")
print("Global Total Deaths:", f"{global_deaths:,.0f}")
```
